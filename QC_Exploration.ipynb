```python
import scanpy as sc, matplotlib.pyplot as plt, seaborn as sns, numpy as np
```

```python
sns.set(context="notebook", style="whitegrid")
```

```python
adata = sc.read_h5ad("bipolar_raw.h5ad")
```

```python
adata
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 4))
sns.histplot(adata.obs['nFeature_RNA'], bins=100, ax=axs[0])
sns.histplot(adata.obs['nCount_RNA'], bins=100, ax=axs[1])
sns.histplot(adata.obs['percent.mt'], bins=100, ax=axs[2])
axs[0].set_title("nFeature_RNA")
axs[1].set_title("nCount_RNA")
axs[2].set_title("percent.mt")
plt.tight_layout()
plt.show()
```

Note: retinal cells are energy heavy so mt content percentage is set to 10%
